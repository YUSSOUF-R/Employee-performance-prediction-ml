# Visualization of Employee Performance Insights

## Department-wise Performance

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv(r"C:\Users\madha\Desktop\Employee Performance Analysis\data\processed\employee_performance_processed.csv")

sns.barplot(x='EmpDepartment', y='PerformanceRating', data=df)
plt.title("Department-wise Employee Performance")
plt.xticks(rotation=45)
plt.show()
```
