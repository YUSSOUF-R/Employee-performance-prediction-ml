## Drop Unnecessary Columns

```python
# Employee Number is a unique identifier and not useful for prediction
df.drop(columns=['EmpNumber'], inplace=True)
```

```python
# Feature Engineering - HR meaningful features

df["ExperienceRatio"] = (
    df["ExperienceYearsInCurrentRole"] /
    (df["TotalWorkExperienceInYears"] + 1)
)

df["PromotionGap"] = (
    df["YearsSinceLastPromotion"] /
    (df["ExperienceYearsAtThisCompany"] + 1)
)
```

## Encode Categorical Variables

```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include='object').columns
categorical_cols

# Apply Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```
