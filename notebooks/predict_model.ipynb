# Employee Performance Prediction System for INX Future Inc

This notebook demonstrates how the trained machine learning model can be used
to predict employee performance ratings based on input employee attributes.
This system can support hiring and performance evaluation decisions.


## Import Libraries & Load Model

```python
import pandas as pd
import joblib
```

```python
# Load trained model
model = joblib.load("employee_performance_model.pkl")
```

## Load Processed Dataset (For Sample Prediction)

```python
df = pd.read_csv(r"C:\Users\madha\Desktop\Employee Performance Analysis\data\processed\employee_performance_processed.csv")

X = df.drop(columns=['PerformanceRating'])
```

## Predict Performance (Sample Employees)

```python
# Predict performance ratings
predictions = model.predict(X)

df['PredictedPerformanceRating'] = predictions

df[['PredictedPerformanceRating']].head()
```

## Predict for a New Employee (Hiring Use Case)

This approach can be integrated into hiring systems to screen candidates objectively.

```python
# Example new employee data (sample values)
new_employee = X.iloc[[0]]

predicted_rating = model.predict(new_employee)

print("Predicted Performance Rating:", predicted_rating[0])
```
